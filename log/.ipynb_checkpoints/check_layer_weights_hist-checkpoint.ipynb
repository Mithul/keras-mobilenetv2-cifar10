```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import json
from collections import OrderedDict
from models.mobilenetv2_build import MobileNetV2
from utils.plot_layer_weights import LayerWeightsHistgram

import matplotlib
matplotlib.use('Agg') 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# open json config file as dict
with open(os.path.join(os.getcwd(), "model_config.json")) as config_file:
    train_config             = json.load(config_file, object_pairs_hook=OrderedDict) # config for training
    meta_info_config         = train_config["meta_info"]# config for batch_size, classes, epoch, image shape
    compression_config       = train_config["compression_config"] # config for the QC-matrix
```

## build the model from trained weights

```python
model = MobileNetV2(input_shape=[32, 32, 3], include_top=True, alpha=1.0, classes=meta_info_config["nb_classes"], **compression_config)
model.load_weights(os.path.join(os.getcwd(), "model_checkpoint_only_weights.h5"), by_name=True)
```

## plot & save "weights histgrams" for each layers in the model

```python
lwh = LayerWeightsHistgram(model, model_name=os.path.basename(os.path.abspath(".")))
```

```python
lwh.plot_weights_hist(order="save")
```
